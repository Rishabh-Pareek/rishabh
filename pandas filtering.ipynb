Create a new column with values KID,ADULT & OLD based on age

```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/titanic-train.csv.txt', index_col='PassengerId')
titanic_data.head()
```

```python
def f(a):
    if (a > 0) & (a<16):
        return 'Kid'
    elif (a>16) & (a<31) :
        return 'Adult'
    else:
        return 'old'
titanic_data['age_catagree'] =  titanic_data.Age.apply(f)
titanic_data.head()
```
